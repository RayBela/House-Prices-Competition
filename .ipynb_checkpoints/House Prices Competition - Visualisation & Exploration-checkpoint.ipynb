Showing all features in the train dataset illiustrated in graphs

```python
def plot_feature(X):
    y = train.loc[X.index, 'SalePrice']
    plt.scatter(X, y)
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title(X.name)
    
for feature in train.columns:
    X = train[feature].dropna()
    fig , (ax1, ax2) = plt.subplots(1,2, figsize=(10,5), sharex=True)
    plt.subplot(ax1)
    plot_feature(X)
    plt.subplot(ax2)
    if train[feature].dtype == np.dtype('object'):
        sns.countplot(X)
    else:
        sns.distplot(X)
        
plt.show()
```

### Numerical features and correlation between them

Describing the statistical information about numerical features in the train dataset

```python
train.describe()
```

Showing correlation matrix

```python
# getting only numerical values
numeric_features = train.select_dtypes(include=[np.number])
```

```python
# Compute the correlation matrix
corr = numeric_features.corr()
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()
```

Correlaton between two features

```python
correlations = train.corr()
attrs = correlations.iloc[:-1,:-1] # all except target

threshold = 0.5
important_corrs = (attrs[abs(attrs) > threshold][attrs != 1.0]) \
    .unstack().dropna().to_dict()

unique_important_corrs = pd.DataFrame(
    list(set([(tuple(sorted(key)), important_corrs[key]) \
    for key in important_corrs])), 
        columns=['Attribute Pair', 'Correlation'])

    # sorted by absolute value
unique_important_corrs = unique_important_corrs.ix[
    abs(unique_important_corrs['Correlation']).argsort()[::-1]]

unique_important_corrs
```

Showing 5 most correlated features with "SalePrice" and the least 5 correlated features with the target value too

```python
print (corr['SalePrice'].sort_values(ascending=False)[:5], '\n')
print (corr['SalePrice'].sort_values(ascending=False)[-5:])
```

### Features with missing values

Showing all features which have at least one missing value

```python
nulls = pd.DataFrame(train.isnull().sum().sort_values(ascending=False)[:25])
nulls.columns = ['Null Count']
nulls.index.name = 'Feature'
nulls
```

Plotting features with missing values

#### For Train set:

```python
all_data_na = (train.isnull().sum() / len(train)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]

f, ax = plt.subplots(figsize=(15, 5))
plt.xticks(rotation='90')
sns.barplot(x=all_data_na.index, y=all_data_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
plt.show()
```

#### For Test set:

```python
all_data_na = (test.isnull().sum() / len(test)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]

f, ax = plt.subplots(figsize=(15, 5))
plt.xticks(rotation='90')
sns.barplot(x=all_data_na.index, y=all_data_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
plt.show()
```
